```python
import pandas as pd
import openpyxl
import numpy as np
import sqlite3 
```

```python
def delete_rows_by_index(file_path, sheet_name, row_indices):
  """Deletes rows by their indices from an Excel file.

  Args:
    file_path: Path to the Excel file.
    sheet_name: Name of the sheet.
    row_indices: A list of row indices to delete (1-based).
  """

  workbook = openpyxl.load_workbook(file_path)
  worksheet = workbook[sheet_name]

  # Delete rows in descending order to avoid index shifting
  for row_index in sorted(row_indices, reverse=True):
    worksheet.delete_rows(row_index)

  workbook.save(file_path)
```

```python
file_path = 'BIDV.xlsx'
sheet_name = 'Sheet1'
rows_to_delete = np.arange(1,13)
print(rows_to_delete)
delete_rows_by_index(file_path, sheet_name, rows_to_delete)
```
